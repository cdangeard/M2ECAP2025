A l'aide du jeu de données préparés dans les TP précédents, nous allons explorer deux méthodes d'optimisation des hyperparamètres : la recherche exhaustive (Grid Search) et la recherche aléatoire (Random Search).

## Optimisation d'une random forest


Analysez les résultats obtenus avec GridSearchCV pour une RandomForestClassifier. (nottament cv_results_ et best_params_), à l'aide de graphiques vous pourrez tirer des conclusions sur l'impact des hyperparamètres et étendre votre grid search si besoin.

Optimisez les autres modèles du TP récédents et comparez les performances obtenues avec les hyperparamètres par défaut et ceux optimisés.
